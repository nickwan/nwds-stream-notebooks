```python
import os
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame()
```

```python
_d = {
    'player':'Johnson, Paris',
    'sus_likes':0,
    'personality':7,
    'fit_check':2,
    'cringe_factor':0,
    'is_him':0,
    'espn':12,
    'pff':16,
    'gtm':10,
}
```

```python
df = pd.concat([df,pd.DataFrame.from_dict([_d])])
df
```

```python
df = df.reset_index(drop=True)
```

```python
df['vibe_score'] = df.loc[:, ['sus_likes', 'personality', 'fit_check', 'cringe_factor']].sum(axis=1)
```

```python
df.sort_values('vibe_score', ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.regplot(data=df, x='espn', y='vibe_score')
plt.show()

sns.regplot(data=df, x='pff', y='vibe_score')
plt.show()

sns.regplot(data=df, x='gtm', y='vibe_score')
plt.show()
```
